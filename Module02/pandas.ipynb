```python
import pandas as pd

# Loading CSV
df = pd.read_csv("students.csv")

print("Original data:")
print(df)
```

```python
# Drop rows with any missing values
df_clean = df.dropna()

print("After dropping missing values:")
print(df_clean)
```

```python
# Summarize
print("Summary statistics:")
print(df_clean.describe())
```

```python
# Summarize
print("Summary statistics:")
print(df_clean.describe(include="all"))
```
